## 6. Subplots Personalizados

Matplotlib te permite organizar gráficos en una cuadrícula utilizando `plt.subplot()`, pero también puedes utilizar `plt.subplots()` para mayor control sobre el diseño de los gráficos.

```python
import matplotlib.pyplot as plt
import numpy as np

# Subplots personalizados
fig, axs = plt.subplots(2, 2, figsize=(10, 6))  # Crear una cuadrícula de 2x2 subplots

# Primer subplot
x = np.linspace(0, 10, 100)
axs[0, 0].plot(x, np.sin(x), color='blue')
axs[0, 0].set_title('Seno')

# Segundo subplot
axs[0, 1].plot(x, np.cos(x), color='green')
axs[0, 1].set_title('Coseno')

# Tercer subplot
axs[1, 0].plot(x, np.tan(x), color='red')
axs[1, 0].set_title('Tangente')

# Cuarto subplot
axs[1, 1].plot(x, np.log(x + 1), color='purple')
axs[1, 1].set_title('Logaritmo')

plt.tight_layout()
plt.show()
```

Aquí hemos creado una cuadrícula de 2x2 con diferentes gráficos en cada celda. La función `plt.tight_layout()` ajusta el espacio entre los subplots para evitar solapamientos.

## 7. Gráfico de Torta (Pie Chart)

Los gráficos de torta son útiles para representar proporciones de un conjunto de datos. Aquí tienes un ejemplo básico de un gráfico de torta en Matplotlib.

```python
# Gráfico de torta
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # Resaltar la primera porción

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Gráfico de Torta')
plt.axis('equal')  # Asegurar que el gráfico sea un círculo
plt.show()
```

En este ejemplo, hemos resaltado la primera porción del gráfico con `explode` y usamos `autopct` para mostrar los porcentajes de cada sección.

## 8. Histogramas

Un histograma es útil para representar la distribución de un conjunto de datos. Matplotlib te permite crear histogramas fácilmente utilizando `plt.hist()`.

```python
# Histograma
data = np.random.randn(1000)
plt.hist(data, bins=30, color='blue', alpha=0.7)
plt.title('Histograma')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

En este ejemplo, hemos generado un conjunto de datos aleatorios utilizando `np.random.randn()` y creado un histograma con 30 barras (`bins=30`).

## 9. Gráficos de Caja (Boxplot)

Los gráficos de caja (boxplots) son útiles para representar la distribución de los datos en términos de cuartiles y mostrar valores atípicos. Vamos a generar un gráfico de caja simple.

```python
# Gráfico de Caja
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data, patch_artist=True)
plt.title('Gráfico de Caja')
plt.xlabel('Conjuntos de Datos')
plt.ylabel('Valores')
plt.grid(True)
plt.show()
```

En este ejemplo, hemos generado tres conjuntos de datos con diferentes desviaciones estándar y los hemos representado en un gráfico de caja para visualizar su distribución.

## 10. Gráficos 3D

Matplotlib también permite crear gráficos tridimensionales utilizando `Axes3D` de la librería `mpl_toolkits.mplot3d`. Vamos a crear un gráfico 3D básico.

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Datos para el gráfico 3D
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Crear superficie 3D
ax.plot_surface(x, y, z, cmap='viridis')
ax.set_title('Superficie 3D')
plt.show()
```

En este gráfico tridimensional, utilizamos `plot_surface` para crear una superficie 3D basada en la función seno de una distancia radial.
