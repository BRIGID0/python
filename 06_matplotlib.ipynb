# Tutorial de Matplotlib para Principiantes

Matplotlib es una de las bibliotecas más utilizadas en Python para la creación de gráficos. Es flexible y permite crear gráficos simples y complejos de manera eficiente.

En este tutorial aprenderás a crear gráficos básicos utilizando Matplotlib.

## 1. Instalación y Primeros Pasos

Si aún no tienes Matplotlib instalado, puedes instalarlo utilizando pip:
```bash
pip install matplotlib
```

Para empezar a utilizar Matplotlib, lo primero que debes hacer es importarlo en tu código de la siguiente manera:

````python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear un gráfico simple
plt.plot(x, y)
plt.title('Gráfico de Seno')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
````

Este código genera un gráfico de la función seno usando valores de `x` entre 0 y 10.

## 2. Personalización de Gráficos

Puedes personalizar fácilmente tus gráficos cambiando colores, estilos de líneas, y añadiendo etiquetas. Aquí tienes un ejemplo:

````python
# Gráfico personalizado
plt.plot(x, y, color='green', linestyle='--', marker='o', label='Seno')
plt.title('Gráfico Personalizado')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()  # Añadir una leyenda
plt.grid(True)  # Añadir una cuadrícula
plt.show()
````

En este caso, hemos añadido una línea verde, punteada (`'--'`), con marcadores circulares (`'o'`) y una cuadrícula.

## 3. Múltiples Gráficos en la Misma Figura

Matplotlib te permite dibujar múltiples gráficos en una sola figura utilizando la función `subplot`.

````python
# Crear una figura con dos gráficos
plt.figure(figsize=(10, 4))

# Primer gráfico
plt.subplot(1, 2, 1)  # 1 fila, 2 columnas, 1er gráfico
plt.plot(x, np.sin(x))
plt.title('Seno')

# Segundo gráfico
plt.subplot(1, 2, 2)  # 1 fila, 2 columnas, 2do gráfico
plt.plot(x, np.cos(x), color='red')
plt.title('Coseno')

plt.tight_layout()
plt.show()
````

En este ejemplo, hemos creado una figura con dos gráficos: el primero muestra el seno de `x` y el segundo muestra el coseno de `x`, organizados uno al lado del otro.

## 4. Guardar Gráficos

Puedes guardar los gráficos generados directamente como archivos de imagen usando el método `savefig`.

````python
# Guardar un gráfico como archivo PNG
plt.plot(x, y)
plt.title('Gráfico de Seno')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.savefig('grafico_seno.png')  # Guardar la figura como PNG
plt.show()
````

## 5. Tipos de Gráficos

Matplotlib permite crear varios tipos de gráficos como barras, histogramas, dispersión, entre otros.

### 5.1 Gráfico de Barras

````python
# Gráfico de barras
categorias = ['A', 'B', 'C', 'D']
valores = [3, 7, 5, 4]

plt.bar(categorias, valores, color='purple')
plt.title('Gráfico de Barras')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()
````

### 5.2 Gráfico de Dispersión

````python
# Gráfico de dispersión
x = np.random.rand(50)
y = np.random.rand(50)
tamaño = np.random.rand(50) * 1000
colores = np.random.rand(50)

plt.scatter(x, y, s=tamaño, c=colores, alpha=0.5)
plt.title('Gráfico de Dispersión')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````
