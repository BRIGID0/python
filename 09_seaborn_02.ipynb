# Tutorial Avanzado de Seaborn

En este tutorial, vamos a explorar características más avanzadas de Seaborn, como la personalización de gráficos, el uso de paletas de colores, facetas, y funciones más avanzadas de gráficos.

## 1. Personalización de Gráficos

Seaborn permite modificar el estilo y la apariencia de los gráficos para hacerlos más atractivos. Vamos a ver cómo modificar los estilos y los contextos.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cambiar el estilo del gráfico
sns.set_style('whitegrid')

# Cambiar el contexto para presentación (gráficos más grandes)
sns.set_context('talk')

# Cargar dataset de ejemplo
data = sns.load_dataset('tips')

# Crear un gráfico de dispersión con estilo personalizado
sns.scatterplot(x='total_bill', y='tip', data=data)
plt.title('Relación entre Total Bill y Tip')
plt.show()
```

Aquí hemos cambiado el estilo del gráfico a `whitegrid` y el contexto a `talk`, que es ideal para presentaciones. Seaborn ofrece otros estilos como `darkgrid`, `white`, y `ticks`.

## 2. Paletas de Colores

Seaborn ofrece varias paletas de colores que puedes utilizar para hacer que tus gráficos se vean más atractivos. También es posible crear paletas personalizadas.

```python
# Usar una paleta de colores predefinida
sns.set_palette('husl')

# Crear un gráfico de barras con la paleta 'husl'
sns.barplot(x='day', y='total_bill', data=data)
plt.title('Total Bill por Día con paleta husl')
plt.show()
```

Seaborn tiene varias paletas predefinidas como `deep`, `muted`, `bright`, `dark`, y `colorblind`. Además, puedes crear tus propias paletas con `sns.color_palette()`.

## 3. Gráficos con Facetas (FacetGrid)

Los gráficos de facetas permiten crear múltiples gráficos a la vez, divididos por categorías. Esto es útil cuando se quiere observar las relaciones entre varias subcategorías de datos.

```python
# Crear un grid de facetas para visualizar múltiples gráficos
g = sns.FacetGrid(data, col='sex', row='time')
g.map(sns.scatterplot, 'total_bill', 'tip')
plt.show()
```

Este ejemplo muestra gráficos de dispersión separados por género (`sex`) y por el momento del día (`time`). La función `FacetGrid` permite crear gráficos más complejos distribuidos en múltiples facetas.

## 4. Gráficos de Regresión Avanzados

Seaborn ofrece una función llamada `regplot` que añade una línea de regresión a los gráficos de dispersión. También puedes ajustar otros parámetros, como la dispersión de los datos y el orden del ajuste polinomial.

```python
# Gráfico de regresión con ajuste polinomial de segundo orden
sns.regplot(x='total_bill', y='tip', data=data, order=2)
plt.title('Regresión polinomial de segundo orden')
plt.show()
```

En este gráfico, hemos utilizado un ajuste polinomial de segundo orden para capturar mejor la relación entre las variables.

## 5. Gráficos de Categorías Avanzados (catplot)

La función `catplot` permite crear gráficos categóricos de manera más flexible, permitiendo cambiar entre diferentes tipos de gráficos fácilmente.

```python
# Crear un gráfico categórico usando catplot
sns.catplot(x='day', y='total_bill', hue='sex', kind='violin', data=data)
plt.title('Distribución de Total Bill por Día y Sexo')
plt.show()
```

Aquí hemos utilizado `catplot` para crear un gráfico de violín, que muestra la distribución de las cuentas (`total_bill`) por día y diferenciadas por género (`sex`). `catplot` es muy versátil y te permite crear gráficos como `boxplot`, `violinplot`, `stripplot`, etc.

## 6. Gráficos de Distribución Multivariada (jointplot y pairplot)

`jointplot` combina un gráfico de dispersión con histogramas para visualizar la distribución de dos variables. `pairplot` muestra múltiples gráficos de dispersión y distribuciones para todas las combinaciones de variables numéricas.

```python
# Gráfico de distribución multivariada con jointplot
sns.jointplot(x='total_bill', y='tip', kind='hex', data=data)
plt.title('Distribución conjunta de Total Bill y Tip')
plt.show()
```

```python
# Gráfico de pares (pairplot)
sns.pairplot(data, hue='sex')
plt.show()
```

El gráfico `jointplot` con `kind='hex'` crea un gráfico de celdas hexagonales para visualizar la densidad de los puntos. `pairplot` genera gráficos de dispersión y distribuciones para todas las combinaciones de variables numéricas, diferenciando por el parámetro `hue`.
